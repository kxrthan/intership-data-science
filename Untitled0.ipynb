```python

import base64


def image_to_binary(image_path):
    try:

        with open(image_path, "rb") as image_file:

            image_binary = image_file.read()


        image_binary_str = ''.join(format(byte, '08b') for byte in image_binary)

        return image_binary_str

    except FileNotFoundError:
        print(f"Error: The file '{image_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {str(e)}")


image_path = "keerthan.jpeg"


binary_data = image_to_binary(image_path)


if binary_data:
    print("Binary data of the image:")
    print(binary_data)
```

```python

import base64
from PIL import Image
import io

def binary_to_image(binary_data, image_path):
    try:
        # Convert the binary string back to binary data
        image_binary = bytearray(int(binary_data[i:i+8], 2) for i in range(0, len(binary_data), 8))

        # Create an image from the binary data
        image = Image.open(io.BytesIO(image_binary))
        image.save(image_path)

        print(f"Image saved as {image_path}")

    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Specify the path to save the output image
output_image_path = "output_keerthan.jpeg"  # Change this to the desired output path

# Specify the binary data obtained from the previous step
binary_data = "
"  # Replace with your binary data

# Call the function to convert binary data to an image
binary_to_image(binary_data, output_image_path)
```
