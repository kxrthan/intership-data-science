```python
import numpy as np
from scipy.stats import entropy

# Hypothetical dataset (attribute values and corresponding class labels)
# You can replace this with your own dataset
attributes = np.array([1, 1, 0, 0, 1, 0, 1, 1, 0, 0])
labels = np.array(["spam Indicator"])

# Function to calculate entropy
def calculate_entropy(labels):
    # Count the occurrences of each class label
    unique_labels, counts = np.unique(labels, return_counts=True)
    # Calculate the probability of each class
    probabilities = counts / len(labels)
    # Calculate entropy using the formula: -p * log2(p)
    entropy_value = -np.sum(probabilities * np.log2(probabilities))
    return entropy_value

# Function to calculate Information Gain for a given attribute
def calculate_information_gain(attribute, labels):
    # Calculate the entropy of the entire dataset
    total_entropy = calculate_entropy(labels)

    # Split the dataset into subsets based on the attribute's values
    attribute_values = np.unique(attribute)
    subset_entropies = []

    for value in attribute_values:
        # Create a subset of labels that corresponds to the current attribute value
        subset_labels = labels[attribute == value]
        # Calculate the entropy of the subset
        subset_entropy = calculate_entropy(subset_labels)
        # Calculate the weighted average of subset entropies
        weight = len(subset_labels) / len(labels)
        weighted_subset_entropy = weight * subset_entropy
        subset_entropies.append(weighted_subset_entropy)

    # Calculate the Information Gain
    information_gain = total_entropy - np.sum(subset_entropies)

    return information_gain

# Calculate Information Gain for each attribute
attribute_names = ['Attachments', 'Link Count', 'Word Count', 'Uppercase Count', 'Exclamation Count', 'Question Count', 'Dollar Count', 'Punctuation Count', 'Spam Indicator']
for i, attribute in enumerate(attribute_names):
    information_gain = calculate_information_gain(attributes, labels)
    print(f"Information Gain for {attribute}: {information_gain:.4f}")
```
